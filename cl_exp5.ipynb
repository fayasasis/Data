```python
# ============================================================
# EXPERIMENT 5 – WORD EMBEDDINGS (WORD2VEC + GLOVE)
# ============================================================

# ------------------------------------------------------------
# INSTALL (Run once in terminal)
# pip install gensim numpy scipy
# ------------------------------------------------------------

import numpy as np
import gensim
from gensim.models import KeyedVectors
from numpy.linalg import norm

# ============================================================
# PART 1 – LOAD WORD2VEC GOOGLE NEWS MODEL
# ============================================================

def load_word2vec_model(path):
    print("\nLoading Word2Vec Model (Google News)...")
    model = KeyedVectors.load_word2vec_format(path, binary=True)
    print("Word2Vec Model Loaded Successfully!")
    return model


# ============================================================
# PART 2 – LOAD GLOVE MODEL
# ============================================================

def load_glove_model(glove_path):

    print("\nLoading GloVe Model...")

    glove_dict = {}

    with open(glove_path, "r", encoding="utf8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype="float32")
            glove_dict[word] = vector

    print("GloVe Model Loaded Successfully!")
    return glove_dict


# ============================================================
# PART 3 – WORD VECTOR DISPLAY
# ============================================================

def display_word_vector(model, word, model_name="Word2Vec"):

    print(f"\n--- {model_name} Vector Info ---")

    try:
        vector = model[word]
        print(f"Word: {word}")
        print(f"Vector Dimension: {len(vector)}")
        print("First 10 values:", vector[:10])

    except:
        print(f"Word '{word}' not found in {model_name} vocabulary")


# ============================================================
# PART 4 – COSINE SIMILARITY FUNCTION
# ============================================================

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (norm(v1) * norm(v2))


# ============================================================
# PART 5 – WORD SIMILARITY CHECK
# ============================================================

def check_similarity_word2vec(model, word1, word2):

    print(f"\nWord2Vec Similarity ({word1}, {word2})")

    try:
        score = model.similarity(word1, word2)
        print("Similarity Score:", score)
    except:
        print("One of the words not in vocabulary")


def check_similarity_glove(model, word1, word2):

    print(f"\nGloVe Similarity ({word1}, {word2})")

    try:
        v1 = model[word1]
        v2 = model[word2]
        score = cosine_similarity(v1, v2)
        print("Similarity Score:", score)
    except:
        print("One of the words not in vocabulary")


# ============================================================
# PART 6 – TOP N SIMILAR WORDS
# ============================================================

def top_similar_word2vec(model, word, topn=5):

    print(f"\nTop {topn} similar words (Word2Vec) for '{word}'")

    try:
        words = model.most_similar(word, topn=topn)
        for w, score in words:
            print(w, "->", score)
    except:
        print("Word not found in vocabulary")


def top_similar_glove(model, word, topn=5):

    print(f"\nTop {topn} similar words (GloVe) for '{word}'")

    try:
        if word not in model:
            print("Word not found in vocabulary")
            return

        target_vec = model[word]
        similarities = {}

        for w in model:
            similarities[w] = cosine_similarity(target_vec, model[w])

        sorted_words = sorted(similarities.items(),
                              key=lambda x: x[1],
                              reverse=True)

        for w, score in sorted_words[1:topn+1]:
            print(w, "->", score)

    except:
        print("Error computing similarity")


# ============================================================
# PART 7 – OOV HANDLING
# ============================================================

def check_oov(model, word, model_name="Model"):

    print(f"\nChecking OOV in {model_name}: {word}")

    try:
        if word in model:
            print("Word exists in vocabulary")
        else:
            print("OOV Word!")
    except:
        print("Error checking OOV")


# ============================================================
# MAIN EXECUTION
# ============================================================

if __name__ == "__main__":

    # --------------------------------------------------------
    # ⚠️ CHANGE PATHS BEFORE RUNNING
    # --------------------------------------------------------

    WORD2VEC_PATH = r"C:\Users\prana\Downloads\CL EXPREIMENTS\GoogleNews-vectors-negative300.bin"
    GLOVE_PATH = r"C:\Users\prana\Downloads\CL EXPREIMENTS\glove.6B.100d.txt"

    # Load Models
    # Comment if not downloaded
    w2v_model = load_word2vec_model(WORD2VEC_PATH)
    glove_model = load_glove_model(GLOVE_PATH)

    print("\nNOTE:")
    print("Uncomment model loading lines after downloading embeddings.")

    # --------------------------------------------------------
    # TEST WORDS
    # --------------------------------------------------------

    word_pairs = [
        ("king", "queen"),
        ("computer", "keyboard"),
        ("doctor", "hospital"),
        ("tree", "laptop")
    ]

    query_words = ["science", "education", "artificial"]

    # --------------------------------------------------------
    # WORD VECTOR DISPLAY EXAMPLE
    # --------------------------------------------------------

    display_word_vector(w2v_model, "king", "Word2Vec")
    display_word_vector(glove_model, "king", "GloVe")

    # --------------------------------------------------------
    # SIMILARITY CHECK
    # --------------------------------------------------------

    for w1, w2 in word_pairs:
         check_similarity_word2vec(w2v_model, w1, w2)
         check_similarity_glove(glove_model, w1, w2)

    # --------------------------------------------------------
    # TOP N SIMILAR WORDS
    # --------------------------------------------------------

    for word in query_words:
         top_similar_word2vec(w2v_model, word, topn=5)
         top_similar_glove(glove_model, word, topn=5)

    # --------------------------------------------------------
    # OOV CHECK
    # --------------------------------------------------------

    check_oov(w2v_model, "chatgpt", "Word2Vec")
    check_oov(glove_model, "chatgpt", "GloVe")

    print("\nExperiment 5 Code Ready!")
```
