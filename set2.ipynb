2. Perform complete preprocessing
pipeline.

```python
# Converting to list for preprocessing
documents = docs['review'].tolist()
pos_documents = pos['review'].tolist()
neg_documents = neg['review'].tolist()
print(len(documents))
print(len(pos_documents))
print(len(neg_documents))
```

```python
# Preprocess


import nltk
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nlp = spacy.load("en_core_web_sm")

stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

results = []

for doc_text in documents:

    # Sentence Segmentation
    sentences = sent_tokenize(doc_text)

    # Tokenization
    tokens = word_tokenize(doc_text)

    # Stopword Removal
    filtered = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]

    # Stemming
    stemmed = [ps.stem(w) for w in filtered]

    # Lemmatization
    lemmatized = [lemmatizer.lemmatize(w) for w in filtered]

    results.append({
        "sentences": sentences,
        "tokens": tokens,
        "filtered": filtered,
        "stemmed": stemmed,
        "lemmatized": lemmatized
    })

print("Processed all 20 documents successfully!")
print(len(results))

Result = pd.DataFrame(results)
Result
```

3. Compare:
* Stemmed vs Lemmatized
* vocabulary size.

```python
import pandas as pd

all_original = []
all_stemmed = []
all_lemmatized = []

for r in results:
    all_original.extend(r["filtered"])
    all_stemmed.extend(r["stemmed"])
    all_lemmatized.extend(r["lemmatized"])

comparison_df = pd.DataFrame({
    "Original": all_original,
    "Stemmed": all_stemmed,
    "Lemmatized": all_lemmatized
})

print(comparison_df.head(30))
```

Perform POS tagging using:

*  NLTK
* spaCy

```python
# nltk
all_nltk_pos = []

for r in results:
    pos_tags = nltk.pos_tag(r["filtered"])
    all_nltk_pos.extend(pos_tags)


nltk_pos = pd.DataFrame(all_nltk_pos, columns=["Word", "POS"])
print("NLTK POS tagging")
print(nltk_pos.head(20))
print(f"\n\n\n\n")

#Spacy

all_spacy_pos = []

for doc_text in documents:
    doc = nlp(doc_text)
    spacy_tags = [(token.text, token.pos_) for token in doc]
    all_spacy_pos.extend(spacy_tags)

nltk_pos = pd.DataFrame(all_nltk_pos, columns=["Word", "POS"])
print("Spacy POS tagging")
print(nltk_pos.head(20))

```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
import string

nltk.download('punkt')
nltk.download('wordnet')

# Combine all text
all_text = " ".join(docs['review'].astype(str))

# Tokenize
tokens = word_tokenize(all_text.lower())

# Remove punctuation
tokens = [w for w in tokens if w.isalpha()]

# Unique words only
unique_words = list(set(tokens))

# Find ambiguous words (more than 1 synset)
ambiguous_words = []

for word in unique_words:
    synsets = wn.synsets(word)
    if len(synsets) > 1:
        ambiguous_words.append((word, len(synsets)))

# Sort by number of meanings
ambiguous_words = sorted(ambiguous_words, key=lambda x: x[1], reverse=True)

# Show top 10 most ambiguous words
print("Top ambiguous words found in your dataset:\n")
for word, count in ambiguous_words[:10]:
    print(word, "->", count, "meanings")
```

```python
word = ambiguous_words[0][0]  # Most ambiguous word
synsets = wn.synsets(word)

print(f"\nSelected Word: {word}\n")

for i, syn in enumerate(synsets[:2]):  # Show first 2 meanings
    print(f"Meaning {i+1}:")
    print("Definition:", syn.definition())
    print("Example:", syn.examples())
    print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=1000
)

tfidf_matrix = vectorizer.fit_transform(documents)
pos_tfidf_matrix = vectorizer.transform(pos_documents)
neg_tfidf_matrix = vectorizer.transform(neg_documents)

print("TF-IDF Shape:", tfidf_matrix.shape)

feature_names = vectorizer.get_feature_names_out()

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names
)

print(tfidf_df.head())
```

TOP 10 pos & neg

```python
import numpy as np

# Average TF-IDF score for positive reviews
pos_mean = np.mean(pos_tfidf_matrix.toarray(), axis=0)

# Average TF-IDF score for negative reviews
neg_mean = np.mean(neg_tfidf_matrix.toarray(), axis=0)

# Get top 10 indices
top_pos_indices = pos_mean.argsort()[-10:][::-1]
top_neg_indices = neg_mean.argsort()[-10:][::-1]

print("Top 10 Positive Words:")
for idx in top_pos_indices:
    print(feature_names[idx], ":", pos_mean[idx])

print("\nTop 10 Negative Words:")
for idx in top_neg_indices:
    print(feature_names[idx], ":", neg_mean[idx])
```

2.Cosine similarity

*   two positve reviews
*   two negative reviews



```python
from sklearn.metrics.pairwise import cosine_similarity

# Take first two positive reviews
pos_vec1 = pos_tfidf_matrix[0]
pos_vec2 = pos_tfidf_matrix[1]

# Compute cosine similarity
similarity_pos_pos = cosine_similarity(pos_vec1, pos_vec2)

print("Cosine Similarity (Positive vs Positive):", similarity_pos_pos[0][0])

# Taking one neg and one pos
pos_vec = pos_tfidf_matrix[0]
neg_vec = neg_tfidf_matrix[0]

similarity_pos_neg = cosine_similarity(pos_vec, neg_vec)

print("Cosine Similarity (Positive vs Negative):", similarity_pos_neg[0][0])
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

x = docs['review']
y = docs['sentiment']



le = LabelEncoder()
y = le.fit_transform(docs['sentiment'])


X_train, X_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)

X_test_text = X_test
# Applying Tfid to train and test
X_train = vectorizer.fit_transform(X_train)
y_train = y_train

X_test = vectorizer.transform(X_test)
y_test = y_test

```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score,classification_report



nb = MultinomialNB()
nb.fit(X_train, y_train)

nb_pred = nb.predict(X_test)

print(classification_report(y_test,nb_pred))


nb_accuracy = accuracy_score(y_test, nb_pred)
nb_precision = precision_score(y_test, nb_pred, average='weighted')
nb_recall = recall_score(y_test, nb_pred, average='weighted')
nb_f1 = f1_score(y_test, nb_pred, average='weighted')
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

print(accuracy_score(y_test, lr_pred))
print(classification_report(y_test,lr_pred))


lr_accuracy = accuracy_score(y_test, lr_pred)
lr_precision = precision_score(y_test, lr_pred, average='weighted')
lr_recall = recall_score(y_test, lr_pred, average='weighted')
lr_f1 = f1_score(y_test, lr_pred, average='weighted')
```

```python
Comparison_table = pd.DataFrame({
    'Model': ['Naive Bayes', 'Logistic Regression'],
    'Accuracy': [nb_accuracy, lr_accuracy],
    'Precision': [nb_precision, lr_accuracy],
    'Recall': [nb_recall, lr_recall],
    'F1 Score': [nb_f1, lr_f1]
})

print(Comparison_table)
```
