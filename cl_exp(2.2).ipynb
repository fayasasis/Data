```python
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv(r"C:\Users\prana\Downloads\NER_Dataset.csv")
```

```python
def str_to_list(s):
    try:
        return eval(s)
    except:
        return []
```

```python
df['Words'] = df['Word'].apply(str_to_list)
df['NER_Tags'] = df['Tag'].apply(str_to_list)
df['NLTK_POS'] = df['POS'].apply(str_to_list)
```

```python
def get_spacy_pos(words):
    doc = nlp(" ".join(words))
    return [token.pos_ for token in doc]
```

```python
doc = Document()
doc.add_heading(
    "POS + NER Comparison (NLTK vs spaCy) - First Sentences",
    level=1
)
```

```python
for _, row in df.head(200).iterrows():
    sentence_id = row['Sentence_ID']
    words = row['Words']
    nltk_pos = row['NLTK_POS']
    ner_tags = row['NER_Tags']
    spacy_pos = get_spacy_pos(words)

    doc.add_paragraph(f"\nSentence ID: {sentence_id}")

    table = doc.add_table(rows=1, cols=4)
    table.style = 'Light List Accent 1'

    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = 'Word'
    hdr_cells[1].text = 'NLTK POS'
    hdr_cells[2].text = 'spaCy POS'
    hdr_cells[3].text = 'NER Tag'
    for w, n_pos, s_pos, ner in zip(words, nltk_pos, spacy_pos, ner_tags):
        row_cells = table.add_row().cells
        row_cells[0].text = str(w)
        row_cells[1].text = str(n_pos)
        row_cells[2].text = str(s_pos)
        row_cells[3].text = str(ner)
```

```python
output_dir = r"C:\Output"
os.makedirs(output_dir, exist_ok=True)
```

```python
file_path = os.path.join(output_dir, "Dataset5_output2.docx")
doc.save(file_path)
```
