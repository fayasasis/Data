```python
import nltk
import spacy
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
datasets = {
    "News Text": (
        "The Reserve Bank of India announced a new monetary policy on Friday. "
        "Governor Shaktikanta Das addressed the media in New Delhi."
    ),

    "Healthcare Text": (
        "The patient was diagnosed with diabetes on 12 March 2024 and "
        "prescribed medication by Dr. Joseph at Apollo Hospital."
    ),

    "Technology Text": (
        "Google introduced a new AI model in 2024 to enhance machine "
        "translation and text understanding capabilities."
    ),

    "Education Text": (
        "Rajagiri School of Engineering and Technology conducted an NLP "
        "workshop for final year students on 10 January 2025 at Kochi campus."
    )
}
```

```python
def print_pos_comparison(text, dataset_name):
    nltk_tokens = nltk.word_tokenize(text)
    nltk_tags = nltk.pos_tag(nltk_tokens)
    spacy_doc = nlp(text)

    print("\nComparison of Outputs")
    print("=" * 80)
    print(f"DATASET: {dataset_name}\n")
    print(f"{'Word':<20} {'NLTK POS':<25} {'spaCy POS'}")
    print("-" * 65)

    min_len = min(len(nltk_tags), len(spacy_doc))
    for i in range(min_len):
        word = nltk_tags[i][0]
        nltk_pos = nltk_tags[i][1]
        spacy_pos = spacy_doc[i].pos_
        print(f"{word:<20} {nltk_pos:<25} {spacy_pos}")
```

```python
for name, text in datasets.items():
    print_pos_comparison(text, name)
```
